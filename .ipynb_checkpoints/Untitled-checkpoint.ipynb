```python
from getpass import getpass
from datetime import datetime
```

```python
class ValidInfo:
    def __init__(self):
        self.valid_domains = ["gmail", "yahoo", "outlook", "hotmail", "icloud", "proton", "aol", "zoho", "rediff", "yandex","gyanadeepvidyalaya"]
        self.valid_extensions = ["com", "in", "org", "net", "edu", "gov", "co", "io", "me", "info"]

    def valid_email(self, email):
        email = email.lower()
        if email.count("@") != 1:
            print("Email should contain exactly one @")
            return False
        username, domain = email.split("@", maxsplit=1)
        if not username:
            print("Username cannot be empty")
            return False
        if not username[0].isalpha():
            print("Email should start with a character")
            return False
        if "." not in domain:
            print("Domain should contain a .")
            return False
        domain_name, extension = domain.split(".", maxsplit=1)
        if domain_name not in self.valid_domains:
            print("Invalid email domain")
            return False
        if extension not in self.valid_extensions:
            print("Invalid email extension")
            return False
        return True

    def get_valid_email(self, prompt):
        while True:
            email = input(prompt)
            if self.valid_email(email):
                return email
            print("Invalid Email. Please try again.")

                        
    def valid_mobile(self, mob):
        if len(mob) == 10 and mob.isdigit():
            return True
        elif len(mob) == 13 and mob.startswith("+") and mob[1:].isdigit():
            return True
        else:
            return False

    def get_valid_mobile(self, prompt):
        while True:
            mob = input(prompt)
            if self.valid_mobile(mob):
                return mob
            print("Invalid mobile number. Please try again.")

    def valid_age(self, age):
        try:
            age = int(age)
            if 0 < age <= 110:
                return True
            else:
                print("Age is not valid")
                return False
                
        except ValueError:
            print("Age should be in number only")
            return False

    def get_valid_age(self, prompt):
        while True:
            age = input(prompt)
            if self.valid_age(age):
                return int(age)
        
    def get_valid_number(self, prompt, data_type=int, min_value=0):
        # Small helper so both Staff and Student managers can safely
        # ask the user for a number without crashing on bad input.
        while True:
            try:
                value = data_type(input(prompt))
                if value < min_value:
                    print(f"Value must be at least {min_value}.")
                    continue
                return value
            except ValueError:
                print(f"Please enter a valid Number.")

    def recheck_password(self, password, confirm_password):
        if password != confirm_password:
            print("Password Not Match")
            return 
        return True

    def get_valid_password(self):
        while True:
            password = getpass("Enter Student's Password :")
            confirm_password = getpass("Re-Enter Student's Password :")
            if self.recheck_password(password,confirm_password):
                return password

    def get_valid_option(self, prompt, options):
        while True:
            choice = input(prompt).strip()
            if choice in options:
                return choice
            print(f"Please enter one of these options: {', '.join(options)}")

    def get_valid_date(self, prompt):
        while True:
            date_of_birth = input(prompt)
            try:
                today = datetime.today().date()
                date_of_birth = datetime.strptime(date_of_birth, "%Y-%m-%d").date()
    
                if date_of_birth > today:
                    print("Date of birth cannot be in the future.")
                    continue
    
                return str(date_of_birth)
    
            except ValueError:
                print("Invalid date format! Please enter the date as YYYY-MM-DD.")
```

```python
class ManageStudent(ValidInfo):
    def AddStudent(self):
        std_id = self.get_valid_number("Enter Student's Id :")
        roll_number = self.get_valid_number("Enter Student's Roll No :")
        first_name = input("Enter Student's First Name:")
        last_name = input("Enter Student's Last Name:")
        gender = gender = self.get_valid_option("Enter Student's Gender (M/F/O): ",["M", "F", "O"])
        date_of_birth = self.get_valid_date("Enter Student's Date of Birth (YYYY-MM-DD): ")
        std_email = self.get_valid_email("Enter Student's Email :")
        age = self.get_valid_age("Enter Student's Age :")
        password = self.get_valid_password()
        phone_number = self.get_valid_mobile("Enter Student's Mobile Number : ")
        address = input("Enter Student's Address : ")
        city = input("Enter Student's City :")
        state = input("Enter Student's State : ")
        pin_code = self.get_valid_number("Enter Student's Pin Code :")
        blood_group = input("Enter Student's Blood Group :")
        std_class = input("Enter Student's Class :")
        section = input("Enter Student's Section :")
        parent_name = input("Enter Student's Parent Name :")
        parent_phone = self.get_valid_mobile("Enter Parent's Mobile Number: ")
        parent_email = self.get_valid_email("Enter Student's Parent Email :")
        admission_date = self.get_valid_date("Enter Student's Admission (YYYY-MM-DD): ")
        fees = self.get_valid_number("Enter Student's Fees :")
        marks = {}
        subject_number = self.get_valid_number("Enter how many Subejct do you want to add :")
        for i in range(subject_number):
            subject_name = input("Enter subject name :")
            subject_marks = self.get_valid_number(f"Enter Marks for: {subject_name} :")
            marks[subject_name] = subject_marks
        status = "active"
        Gyanadeep_Vidyalaya[std_email] = {'id':std_id, 'roll_number':roll_number, 'first_name':first_name, 'last_name':last_name, 'gender':gender, 
                        'date_of_birth':date_of_birth, 'age':age, 'password':password, 'phone_number':phone_number, 'address':address,
                        'city':city, 'state':state, 'pin_code':pin_code, 'blood_group':blood_group, 'class':std_class, 
                        'section':section, 'parent_name':parent_name, 'parent_phone':parent_phone, 'parent_email':parent_email,
                        'admission_date':admission_date, 'fees':fees, 'marks':marks, 'status':status}

    def ViewStudent(self):
        email = self.get_valid_email("Enter Student's Email :")
        if email not in Gyanadeep_Vidyalaya["student"]:
            print("Student Not Found!")
            return
        print({k: v for k, v in Gyanadeep_Vidyalaya['student'][email].items() if k != "password" })

    def ViewAllStudents(self):
        print({email: {k: v for k, v in student.items() if k != "password"}for email, student in Gyanadeep_Vidyalaya["student"].items()})

    def EditStudent(self):
        email = self.get_valid_email("Enter Student Email: ")
        if email not in Gyanadeep_Vidyalaya["student"]:
            print("Student not found.")
            return
        student = Gyanadeep_Vidyalaya["student"][email]
        fields = {
            1: ("id", lambda: self.get_valid_number("Enter New ID: ")),
            2: ("roll_number", lambda: self.get_valid_number("Enter New Roll Number: ")),
            3: ("first_name", lambda: input("Enter New First Name: ")),
            4: ("last_name", lambda: input("Enter New Last Name: ")),
            5: ("gender", lambda: self.get_valid_option("Enter Gender (M/F/O): ", ["M", "F", "O"])),
            6: ("date_of_birth", lambda: self.get_valid_date("Enter Date of Birth: ")),
            7: ("age", lambda: self.get_valid_age("Enter New Age: ")),
            8: ("password", self.get_valid_password),
            9: ("phone_number", lambda: self.get_valid_mobile("Enter New Phone Number: ")),
            10: ("address", lambda: input("Enter New Address: ")),
            11: ("city", lambda: input("Enter New City: ")),
            12: ("state", lambda: input("Enter New State: ")),
            13: ("pin_code", lambda: self.get_valid_number("Enter New Pin Code: ")),
            14: ("blood_group", lambda: input("Enter New Blood Group: ")),
            15: ("class", lambda: input("Enter New Class: ")),
            16: ("section", lambda: input("Enter New Section: ")),
            17: ("parent_name", lambda: input("Enter New Parent Name: ")),
            18: ("parent_phone", lambda: self.get_valid_mobile("Enter New Parent Phone: ")),
            19: ("parent_email", lambda: self.get_valid_email("Enter New Parent Email: ")),
            20: ("admission_date", lambda: self.get_valid_date("Enter Admission Date: ")),
            21: ("fees", lambda: self.get_valid_number("Enter New Fees: ")),
            23: ("status", lambda: self.get_valid_option("Enter Status (active/inactive): ", ["active", "inactive"]))
        }
        menu = [
            "ID", "Roll Number", "First Name", "Last Name", "Gender",
            "Date of Birth", "Age", "Password", "Phone Number", "Address",
            "City", "State", "Pin Code", "Blood Group", "Class",
            "Section", "Parent Name", "Parent Phone", "Parent Email",
            "Admission Date", "Fees", "Marks", "Status"
        ]
        print("\n========== Edit Student ==========")
        for i, item in enumerate(menu, start=1):
            print(f"{i}. {item}")
        choice = self.get_valid_number("Enter Choice: ")
        if choice == 22:
            print("\nCurrent Subjects and Marks")
            print("-" * 30)
            for subject, mark in student["marks"].items():
                print(f"{subject} : {mark}")
    
            print("\n1. Edit Existing Subject")
            print("2. Add New Subject")
            mark_choice = self.get_valid_number("Enter Choice: ")
            if mark_choice == 1:
                subject = input("Enter Subject Name: ")
    
                if subject in student["marks"]:
                    student["marks"][subject] = self.get_valid_number(
                        f"Enter New Marks for {subject}: "
                    )
                    print("Marks Updated Successfully.")
                else:
                    print("Subject not found.")
            elif mark_choice == 2:
                subject = input("Enter New Subject Name: ")
    
                if subject in student["marks"]:
                    print("Subject already exists.")
                else:
                    student["marks"][subject] = self.get_valid_number(
                        f"Enter Marks for {subject}: "
                    )
                    print("New Subject Added Successfully.")
            else:
                print("Invalid Choice.")
        elif choice in fields:
            key, func = fields[choice]
            student[key] = func()
            print("Student information updated successfully.")
        else:
            print("Invalid Choice.")

    def DeleteStudent(self):
        email = self.get_valid_email("Enter Student's Email you want to Delete")
        if email not in Gyanadeep_Vidyalaya['student']:
            print("Student Not Found!")
            return
        student = Gyanadeep_Vidyalaya["student"][email]
        choice = self.get_valid_option(f"Are you sure to delete {student['first_name']+" "+student['last_name']}'s Record ('y'/'n') :",["y","n"])
        if choice =='y':
            Gyanadeep_Vidyalaya_Deleted['student'][email] = Gyanadeep_Vidyalaya['student'].pop(email)
            print(f"Student {student['first_name']+" "+student['last_name']}'s record deleted successfully")
        else:
            print("Deletion cancelled..")
```
